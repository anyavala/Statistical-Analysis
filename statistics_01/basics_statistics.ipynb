## Learning Statistics with Python: Central Tendency and Data Dispersion

```python
#Install the libraries into your environment
#pip install numpy
#pip install scipy
```

```python
# Suppose we have a sample data
data = [2, 4, 6, 8, 10, 4, 7, 3, 5]
```

## 1. Measures of Central Tendency
### Using Pure Python

```python
# Mean
mean = sum(data) / len(data)
```

```python
# Median
data_sorted = sorted(data)
n = len(data)
if n % 2 == 0:
    median = (data_sorted[n//2 - 1] + data_sorted[n//2]) / 2
else:
    median = data_sorted[n//2]
```

```python
# Mode
from collections import Counter
data_count = Counter(data)
mode = [k for k, v in data_count.items() if v == max(data_count.values())]

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

### Using NumPy and SciPy

```python
import numpy as np
from scipy import stats

mean_np = np.mean(data)
median_np = np.median(data)
mode_np = stats.mode(data).mode

print("Mean (NumPy):", mean_np)
print("Median (NumPy):", median_np)
print("Mode (SciPy):", mode_np)
```

## 2. Measures of Dispersion
### Using Pure Python

```python
# Range= Maximum value - Minimum value
range_data = max(data) - min(data)
```

```python
# Variance
variance_population = sum((x - mean)**2 for x in data) / len(data)
variance_sample = sum((x - mean)**2 for x in data) / (len(data)-1)
```

```python
# Standard Deviation
std_population = variance_population**0.5
std_sample = variance_sample**0.5
```

```python
print("Range:", range_data)
print("Population Variance:", variance_population)
print("Sample Variance:", variance_sample)
print("Population SD:", std_population)
print("Sample SD:", std_sample)
```

### Using NumPy

```python
variance_np_population = np.var(data)
variance_np_sample = np.var(data, ddof=1)
std_np_population = np.std(data)
std_np_sample = np.std(data, ddof=1)

print("Population Variance (NumPy):", variance_np_population)
print("Sample Variance (NumPy):", variance_np_sample)
print("Population SD (NumPy):", std_np_population)
print("Sample SD (NumPy):", std_np_sample)
```

### Interquartile Range (IQR) Using NumPy

```python
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("Interquartile Range (IQR):", IQR)
```

After gaining a basic understanding of these concepts, let's apply the same ideas to real-world data using pandas. Seaborn is a Python library for data visualization that includes several built-in example datasets, such as Titanic, Iris, Tips, and Penguins. In this example, we will focus on the Iris dataset, which contains measurements of various iris flower species.

```python
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')
print(iris.head())
```

```python
# Select numeric columns only
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris_numeric = iris[numeric_cols]
```

```python
# Central Tendency; mean, median, mode
mean_df = iris_numeric.mean()
median_df = iris_numeric.median()
mode_df = iris_numeric.mode().iloc[0]  # take the first row in case of multiple modes
print("Mean (DataFrame):\n", mean_df)
print("Median (DataFrame):\n", median_df)
print("Mode (DataFrame):\n", mode_df)
```

```python
# Dispersion
range_df = iris_numeric.max() - iris_numeric.min()
variance_df = iris_numeric.var()      # sample variance by default (ddof=1)
std_df = iris_numeric.std()           # sample standard deviation
iqr_df = iris_numeric.quantile(0.75) - iris_numeric.quantile(0.25)
print("Range (DataFrame):\n", range_df)
print("Variance (DataFrame):\n", variance_df)
print("Standard Deviation (DataFrame):\n", std_df)
print("Interquartile Range (DataFrame):\n", iqr_df)
```

```python
# Combine all statistics into a single DataFrame
import pandas as pd
iris_stats_df = pd.DataFrame({
    'Mean': mean_df,
    'Median': median_df,
    'Mode': mode_df,
    'Range': range_df,
    'Variance': variance_df,
    'Standard Deviation': std_df,
    'IQR': iqr_df
})

# Display the result
print(iris_stats_df)
```

### How to interpret the results?
- Range (max - min): Shows the spread from smallest to largest values. Petal length has the largest range (5.9), while sepal width is least variable (2.4).
- Variance: Measures how spread out values are around the mean. Petal length has the highest variance (3.11), sepal width the lowest (0.19).
- Standard Deviation (SD): Petal_length has SD ≈ 1.76, meaning on average, values deviate ~1.76 units from the mean. Sepal_width SD ≈ 0.44, showing it’s very consistent.
- Interquartile Range (IQR): Petal_length IQR ≈ 3.55 → middle 50% of petal lengths spans 3.55 units.
Sepal_width IQR ≈ 0.5 → middle 50% of sepal widths is tightly packed.
