## T- TEST EXAMPLE

```python
import scipy.stats as stats

# Data
method_A = [85, 88, 90, 92, 87]
method_B = [78, 82, 79, 81, 77]

# Perform independent t-test
t_stat, p_value = stats.ttest_ind(method_A, method_B)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# Significance level
alpha = 0.05

# Decision
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two methods.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the two methods.")
```

## Comparing Two Groups in the Iris Dataset

####  Statistical Hypotheses
- Null hypothesis (H₀):
The mean petal length of setosa and versicolor are equal (μ₁ = μ₂)
- Alternative hypothesis (H₁):
The mean petal lengths are different (μ₁ ≠ μ₂)

```python
# To-sample t-test
import seaborn as sns
from scipy import stats

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Select petal lengths for the two species
setosa = iris[iris['species'] == 'setosa']['petal_length']
versicolor = iris[iris['species'] == 'versicolor']['petal_length']

# Perform independent two‑sample t‑test
t_stat, p_value = stats.ttest_ind(setosa, versicolor)

print(f"T‑statistic: {t_stat:.3f}")
print(f"P‑value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject H₀ → Significant difference in mean petal lengths")
else:
    print("Fail to reject H₀ → No significant difference")
```

```python
import numpy as np
from scipy import stats

# Generate two groups from similar distributions
np.random.seed(1)
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=51, scale=5, size=30)

# Perform independent t-test
t_stat, p_value = stats.ttest_ind(group1, group2)

print(f"T‑statistic: {t_stat:.3f}")
print(f"P‑value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject H₀ → Significant difference in means")
else:
    print("Fail to reject H₀ → No significant difference in means")
```

```python
import pandas as pd
from scipy import stats

# Load the Wine dataset from UCI
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
cols = ['Class', 'Alcohol', 'Malic acid','Ash','Alcalinity of ash','Magnesium',
        'Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins',
        'Color intensity','Hue','OD280/OD315','Proline']
wine = pd.read_csv(url, names=cols)

# Select alcohol values for two classes
group1 = wine[wine['Class'] == 1]['Alcohol']
group2 = wine[wine['Class'] == 2]['Alcohol']

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(group1, group2)

print("Wine Dataset:")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```
